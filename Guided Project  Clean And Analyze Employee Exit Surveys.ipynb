## Identifying Dissatisfied Employees. 

I have created a new institute_service column that I'll use to analyze survey respondents according to their length of employment.Next, I'll identify any employees who resigned because they were dissatisfied.

Below are the columns I'll use to categorize employees as 'dissatisfied' from each dataframe.

tafe_survey_updated:

If the employee indicated any of the factors above caused them to resign, I'll mark them as ``dissatisfied`` in a new column.

To create the new column, I'll do the following:

Convert the values in the ``Contributing Factors. Dissatisfaction`` and ``Contributing Factors. Job Dissatisfaction`` columns in the ``tafe_resignations`` dataframe to ``True``, ``False``, or ``NaN`` values.
If any of the columns listed above contains a ``True`` value, I'll add a ``True`` value to a new column named ``dissatisfied``. To accomplish this, I'll use the ``DataFrame.any()`` method on both dataframes to do the following:

- Return ``True`` if any element in the selected columns above is ``True``
- Return ``False`` if none of the elements in the selected columns above is ``True``
- Return ``NaN`` if the value is ``NaN``

Here's the syntax I can use:``df.any(axis=1, skipna=False)``.After my changes, the new ``dissatisfied`` column will contain just the following values:

- ``True``: indicates a person resigned because they were dissatisfied with the job
- ``False``: indicates a person resigned because of a reason other than dissatisfaction with the job
- ``NaN``: indicates the value is missing

I will then use the ``df.copy()`` method to create a copy of the results and avoid the ``SettingWithCopy`` warning and assign the results to ``dete_resignations_up`` and ``tafe_resignations_up``.


```python
# Checking for unique values
tafe_resignations ['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
# Check the unique values
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
# Update the values in the contributing factors columns to be either True, False, or NaN
def update_vals(x):
    if x == '-':
        return False
    elif pd.isnull(x):
        return np.nan
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()

# Check the unique values after the updates
tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
# Update the values in columns related to dissatisfaction to be either True, False, or NaN
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
combined_updated['institute_service'].value_counts()
```

To analyze the data, I'll convert these numbers into categories. I'll base my analysis on this [article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage), which makes the argument that understanding employee's needs according to career stage instead of age is more effective.

I'll use the slightly modified definitions below:

- New: Less than 3 years at a company
- Experienced: 3-6 years at a company
- Established: 7-10 years at a company
- Veteran: 11 or more years at a company

I'll extract the years of service from each value in the ``institute_service`` column. First I'll change the type to ``'str'``.Then extract the years of service from each pattern. I'll then double check that I didn't miss extracting any digits. Then change the type to ``'float'``.


```python
# Change type and extract the years of service from each pattern.
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
```

```python
# Change to 'float'
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')

# Check if all the values have been extracted correctly 
combined_updated['institute_service_up'].value_counts()
```

Next, I'll map each value to one of the career stage definitions above by creating a function that maps each year value to one of the career stages above.I will use the following code to check if a value is ``NaN`` where val is the name of the value: ``pd.isnull(val)``. I will then use the ``Series.apply()`` method to apply the function to the ``institute_service`` column and assign the result to a new column named ``service_cat``.

```python
# Convert years of service to categories
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)

# Quick check of the update
combined_updated['service_cat'].value_counts()
```

```python
combined_updated .info()
```

## Analysing Institutions.

Earlier in the analysis I created the ``institute`` column for each survey to contain values ``DETE`` and ``TAFE`` respectively. I will now analyze each survey separately. 

```python
# Aggregate dissatisfied values according to each institute column.
# Sum results and change data type to float.
dis_num_dete=dete_resignations_up.pivot_table(index='institute', values=['dissatisfied'], aggfunc='sum').astype(float)
dis_num_tafe=tafe_resignations_up.pivot_table(index='institute', values=['dissatisfied'], aggfunc='sum').astype(float)

# Combine results and plot a table.
vertical_stack = pd.concat([dis_num_dete,dis_num_tafe], axis=0)
print(vertical_stack)
```

```python
#Aggregate combined dissatisfied values according to combined institute. 
inst_num=vertical_stack.pivot_table(index='institute', values='dissatisfied',aggfunc='sum').astype(float)

# Visualize the results
inst_num.plot(kind='bar',rot=30)
```

The results above show that more employees resigned due to some dissatisfaction from the DETE than from TAFE.
